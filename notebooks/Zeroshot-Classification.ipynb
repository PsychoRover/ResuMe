# Imports

```python
import transformers as tr
import toml
import pandas as pd

from datasets import Dataset
from sklearn.metrics import classification_report
```

# Load data

```python
# Loading classes .toml
classes = toml.load("./classes.toml")["classes"]
```

```python
df = pd.read_excel("/content/ResuMe-Dataset.xlsx")
```

```python
df = df[df["Category"].isin(classes)]
shuffled_df = df.sample(frac=1)
shuffled_df = shuffled_df.reset_index(drop=True)
```

```python
dataset = Dataset.from_pandas(pd.DataFrame({'data': shuffled_df["Resume"]}))
```

# Experiments

## roberta-large-mnli

```python
# https://huggingface.co/roberta-large-mnli

pipe = tr.pipeline(task="zero-shot-classification", model="roberta-large-mnli", device=0)
```

```python
results = pipe(
    dataset["data"],
    candidate_labels=classes,
    batch_size=64
)
```

### Classification Report

```python
top_scores = [max(zip(result['scores'], result["labels"])) for result in results]
```

```python
y_hat = [label for _, label in top_scores]
```

```python
print(classification_report(shuffled_df.Category, y_hat))
```

## mDeBERTa-v3-base-mnli-xnli

```python
# https://huggingface.co/MoritzLaurer/mDeBERTa-v3-base-mnli-xnli

pipe = tr.pipeline(task="zero-shot-classification", model="MoritzLaurer/mDeBERTa-v3-base-mnli-xnli", device=0)
```

```python
results_2 = pipe(
    dataset["data"],
    candidate_labels=classes,
    batch_size=64
)
```

### Classification Report

```python
top_scores = [max(zip(result['scores'], result["labels"])) for result in results_2]
```

```python
y_hat = [label for _, label in top_scores]
```

```python
print(classification_report(shuffled_df.Category, y_hat))
```
